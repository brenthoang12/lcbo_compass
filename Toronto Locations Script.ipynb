```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/liquor_retail_sales_locations_master_data_november_21.csv')
df = df.drop(df.columns[1:10], axis=1)
df.head()

# Filter for data that has 'Toronto' in 'City/Town' and 'LCBO' in 'Location'
toronto_lcbo_df = df[df['City/Town'].str.contains('Toronto', case=False, na=False) & \
                     df['Location'].str.contains('LCBO', case=False, na=False)]

# Display the head of the filtered DataFrame
toronto_lcbo_df.head()
toronto_lcbo_df.to_csv('toronto_lcbo_data.csv', index=False)

```
